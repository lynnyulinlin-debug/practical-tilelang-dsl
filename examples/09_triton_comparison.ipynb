## 一句话判断

```python
decision = {
    'need_quick_experiment': 'Triton',
    'need_cross_platform_migration': 'TileLang',
    'need_explicit_layout_control': 'TileLang',
    'need_minimal_kernel_experiment': 'Triton',
}

for need, choice in decision.items():
    print(f'{need} -> {choice}')
```
