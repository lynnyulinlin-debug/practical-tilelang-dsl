# Chapter 10: Project Report Layer

This notebook isolates the reporting layer of Chapter 10. It shows how to turn a migration run into a reusable project report and benchmark record.

## What this notebook teaches

- How to record project layer, case style, and subgraph label
- How to keep the benchmark record aligned with the chapter narrative
- How to save a report artifact that can be reused in the written summary

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
for candidate in [ROOT, *ROOT.parents]:
    if (candidate / "HANDOFF.md").exists():
        ROOT = candidate
        break
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from pathlib import Path
import json

from examples.ch10_end_to_end_core import ProjectConfig, run_project

config = ProjectConfig(batch=1, seq_len=16, dim=32, block_size=8, runs=3)
targets = ["cuda", "rocm", "ascend"]

print('Chapter 10: Project Report Layer')
print('=' * 80)
print('This layer focuses on packaging the migration run into a report artifact.')
print()

run_project(
    config,
    targets,
    case_style='operator',
    subgraph_label='toy decoder subgraph',
    sweep=True,
    write_report=True,
    report_path='artifacts/ch10_end_to_end_report.json',
)

report_path = Path('artifacts/ch10_end_to_end_report.json')
if report_path.exists():
    report = json.loads(report_path.read_text())
    print()
    print('Report artifact summary:')
    for key in ['case_style', 'subgraph_label', 'sweep_enabled']:
        print(f'  {key}: {report.get(key)}')
    print('  benchmark_record keys:', ', '.join(sorted(report.get('benchmark_record', {}).keys())[:6]), '...')
```

## Takeaway

The report layer turns a runnable migration example into a reusable artifact that can be cited in the chapter text.
