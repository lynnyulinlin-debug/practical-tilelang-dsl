# Chapter 5: Autotuning Focused Notebook

This notebook isolates the autotuning decision in Chapter 5. It shows how the roofline-style judgment turns into a next-step hint.

## What this notebook teaches

- How to turn a profile into a concrete next step
- Why autotuning should come after the bottleneck is known
- How to keep search guidance aligned with engineering judgment

```python
from pathlib import Path
import sys

ROOT = Path.cwd().resolve()
for candidate in [ROOT, *ROOT.parents]:
    if (candidate / "HANDOFF.md").exists():
        ROOT = candidate
        break
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from examples.benchmark_record_template import set_benchmark_record
from examples.ch05_schedule_optimize_core import get_profiles, next_step

peak_compute_tflops = 312.0
peak_memory_gbs = 2000.0

print('Chapter 5: Autotuning Next-Step Hints')
print('=' * 72)
for profile in get_profiles():
    print(f'{profile.name}')
    print(f'  Next: {next_step(profile, peak_compute_tflops, peak_memory_gbs)}')
    print()

record = set_benchmark_record(
    scenario='chapter 5 autotuning hint demo',
    operator='analysis',
    platform='NVIDIA',
    target='cuda',
    dtype='float32',
    shape='analysis only',
    baseline='roofline estimate',
    optimization='autotuning hint',
)
print('Benchmark record skeleton:')
for key in ['scenario', 'operator', 'platform', 'target', 'dtype', 'shape', 'baseline', 'optimization']:
    print(f'  {key}: {record[key]}')
```

## Takeaway

Autotuning should refine a known direction, not replace the direction-finding step.
