# 第 1 章教学讲义：为什么需要 DSL

这个 notebook 用来把第 1 章的判断框架讲清楚：kernel、DSL、compiler、backend 分别负责什么，为什么 AI kernel 的问题会推动抽象层上移。

## 1. 先看问题

- 单个 kernel 可以写出来，不代表整类 kernel 好维护
- 当 shape、硬件、后端变化频繁时，手写实现的边际成本会迅速上升
- DSL 的价值不是“换个写法”，而是把表达、组织和优化变成可分析对象

```python
# 一个极简的判断框架：
kernel = "执行单元"
dsl = "描述和组织 kernel 的方式"
compiler = "把高层意图变成可执行实现的桥梁"

print(kernel)
print(dsl)
print(compiler)
```

## 2. 为什么不是只有 kernel

如果一个问题只服务固定 shape 和固定硬件，手写 kernel 可能够用。  
如果它需要跨 shape、跨后端、跨优化路径，DSL 往往更有长期价值。

## 3. TileLang 的位置

- 它不隐藏硬件细节
- 它把 tile、layout、memory tier 和并行控制显式写出来
- 它让编译链路承接一部分复杂度，但不会替你自动解决所有性能问题

```python
# 关键判断：
needs_dsl = [
    "要覆盖多种 shape",
    "要跨多个后端",
    "要把调优流程复用起来",
    "要保留显式控制能力",
]

for item in needs_dsl:
    print(f"- {item}")
```

## 4. 本章结论

读完这章，读者应该能判断：什么时候继续手写 kernel，什么时候上提到 DSL 层，什么时候仍然必须依赖 benchmark 和 profiler。

## 5. 一个更实用的判断表

| 场景 | 更适合谁 | 原因 |
|---|---|---|
| 单一固定算子 | 手写 kernel | 需求稳定，直接控制最省事 |
| 多种 shape / 多后端 | DSL | 复用和迁移更重要 |
| 需要持续调优 | DSL + 编译器 | 便于把优化流程结构化 |
| 需要追峰值性能 | kernel + profiler | 最终还是要回到实现和数据 |

这张表的目的不是给绝对答案，而是提供一个第一轮判断方向。

```python
cases = [
    ("fixed kernel", "handwritten"),
    ("many shapes", "DSL"),
    ("many backends", "DSL"),
    ("peak tuning", "kernel + profiler"),
]
for name, choice in cases:
    print(f"{name:>14} -> {choice}")
```
